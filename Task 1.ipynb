```python
#here I first Import some of the important libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('stats_sample_data.csv')#Reading the data
```

```python
data.head()
```

```python
data.info()#as we see there is no null values
```

```python
data.sort_values(by='Date',inplace=True)#sorted the data according to the date.
```

```python
data.head()
```

```python
#here the graph is showing that the Runner game is used mostly in differnet countries compairing to the other games.
plt.figure(figsize=(9,12))
sns.countplot(y='Name',data=data)
```

```python
#here I plot the graph which shows that the maximum downloaded game is Bath
plt.figure(figsize=(16,7))
sns.barplot(x='Name',y='Downloads',data=data)
plt.xticks(rotation=45)
plt.show()
```

```python
#here we can see that the maximum viewing page is of the Bath.
plt.figure(figsize=(16,7))
sns.barplot(x='Name',y='Product Page Views',data=data)
plt.xticks(rotation=45)
plt.show
```

```python
#this graph shows the country using or downloading most of the games.
plt.figure(figsize=(16,8))
sns.countplot(x=data['Country'])
plt.xticks(rotation=45)
plt.show()
```

```python
#this graph shows the highest country among the others who download maximum number of games.
plt.figure(figsize=(16,8))
sns.barplot(x=data['Country'],y=data['Downloads'])
plt.xticks(rotation=45)
plt.show()
```

```python
#this graph shows the download done by the country in the 2 year.Where USA is the highest in downloading the game.
plt.figure(figsize=(16,8))
sns.barplot(x=data['Country'],y=data['Downloads'],hue=data['Date'].dt.year)
plt.xticks(rotation=45)
plt.show()
```

```python
#here we split the data according to the years.
data1=data.loc[data['Date'].dt.year==2019]#this is the data for 2019
data2=data.loc[data['Date'].dt.year==2020]#this is the data for 2020
```

```python
data1.head()
```

```python
b=data1.loc[data1['Name']=='Bakery']['Downloads']
s=data1.loc[data1['Name']=='Surf Fun']['Downloads']
```

Compairing two differnet games in the behalf of there downloads.

```python
sns.lineplot(x=data1['Date'].dt.month,y=b)#data of Bakery
sns.lineplot(x=data1['Date'].dt.month,y=s)#data of Surf Fun
```

The above graph is showing the comparision of the games as downloaded per month in 2019

```python
#the most downloaded game in 2019 is Bath in comparision to others
plt.figure(figsize=(15,6))
sns.barplot(x='Name',y='Downloads',data=data1)
plt.xticks(rotation=45)
plt.show()
```

```python
#here is the download of the games according to the product version.
plt.figure(figsize=(16,8))
sns.barplot(x='Name',y='Downloads',hue='Product Version',data=data1)
plt.xticks(rotation=45)
plt.show()
```

```python
#here is the download of the games according to the product version in 2020
plt.figure(figsize=(16,8))
sns.barplot(x='Name',y='Downloads',hue='Product Version',data=data2)
plt.xticks(rotation=45)
plt.show()
```

From the both above graph we can see that the download is increased as the newer version is relesed but also some of the older version are good than the new released.And also some games
